```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score, classification_report
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv('https://github.com/Devesh182/project2024/raw/refs/heads/main/DC%20machine%20predictive%20maintenance%20dataset%20file.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
X = dataset.drop( columns =["Motor_ID","Date","Fault Indicator"])
```

```python
Y = dataset ["Fault Indicator"]
```

```python
scaler = StandardScaler()
```

```python
X_scaled= scaler.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.3, random_state=42)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred= model.predict(X_test)
```

```python
accuracy = accuracy_score( Y_test,y_pred)
```

```python
precision= precision_score(Y_test,y_pred)
```

```python
recall= recall_score(Y_test,y_pred)
```

```python
f1 = f1_score(Y_test,y_pred)
```

```python
classification_rep = classification_report(Y_test,y_pred)
```

```python
print("Accuracy:",accuracy)
```

```python
print("Precision:", precision)
```

```python
print("Recall:", recall)
```

```python
print ("F1 score :", f1)
```

```python
print("\nClassifiaction Report:\n", classification_rep )
```

```python
from sklearn.model_selection import learning_curve
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
train_sizes, train_scores, test_scores = learning_curve(
    model, X_scaled, Y, cv=5, scoring='accuracy', n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10), random_state=42)
```

```python
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores_mean, 'o-', color="blue", label="Training score")
plt.plot(train_sizes, test_scores_mean, 'o-', color="green", label="Cross-validation score")
```

```python
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.1, color="blue")
plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha=0.1, color="green")
```

```python
plt.title("Learning Curve (Random Forest Classifier)")
plt.xlabel("Training Set Size")
plt.ylabel("Accuracy")
plt.legend(loc="best")
plt.grid()
plt.show()
```
